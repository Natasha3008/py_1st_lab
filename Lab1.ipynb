Задание 3. Создать двумерные карты с траектроиями из предыдущего задания, используя библиотеку matplotlib. Подписать оси, размер картинки 8*8, не менее 1000 точек.

```python
#код взят с прошлого задания с некоторыми изменениями
import matplotlib.pyplot as plt
import numpy as np

def determination_of_next_points(a, b, x, y): #функция для вычисления значений новых координат на основании текущих значений
  x_new = 1.0 - a * x ** 2.0 + y
  y_new = b * x
  return x_new, y_new

parametrs = [(0.8, 0.3), (1.25, 0.04), (1.42, 0.26)]
num = 1000 ## 2) увеличила количество итераций до 1000
plt.figure(figsize = (8, 8)) ## 1) добавила размер 8*8
for a, b in parametrs:
  x = 0.0    #координаты начальной точки
  y = 0.0
  xPoints = [x]   #список для хранения точек х
  yPoints = [y]   #список для хранения точек у
  for i in range (num):   #в цикле находим значения всех точек
    x, y = determination_of_next_points(a, b, x, y)
    xPoints.append(x)  #добавляем в конец списка
    yPoints.append(y)
  plt.figure()
  plt.grid()
  plt.plot(xPoints, yPoints)
  plt.xlabel('X')  # 3) подписала оси X и Y
  plt.ylabel('Y')
  plt.title(f'a={a}, b={b}')
plt.show()
```

Задание 4. Сделать как в двух предыдущих заданиях, но для трехмерных систем и карты должны быть трехмерные.

B = -0.5 (C, A): (0.13, 0.61); (-0.89, -0.12); (-1.38, -0.42)

     x' = y
     y' = z
     z' = B * x + C * y + A * z - y ** 2.0

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def determination_of_next_points(a, b, c, x, y, z):
  x_new = y
  y_new = z
  z_new = b * x + c * y + a * z - y ** 2.0
  return x_new, y_new, z_new

parametrs = [(0.61, -0.5, 0.13), (-0.12, -0.5, -0.89), (-0.42, -0.5, -1.38)]
num = 1000

for a, b, c in parametrs:
  plt.figure(figsize=(8, 8))
  ax = plt.axes(projection='3d')
  x = 0.001    #координаты начальной точки
  y = 0.056
  z = 0.01
  xPoints = [x]   #список для хранения точек х
  yPoints = [y]   #список для хранения точек у
  zPoints = [z]
  for i in range (num):   #в цикле находим значения всех точек
    x, y, z = determination_of_next_points(a, b, c, x, y, z)
    xPoints.append(x)  #добавляем в конец списка
    yPoints.append(y)
    zPoints.append(z)
  plt.grid()
  ax.plot(xPoints, yPoints, zPoints)
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_zlabel('Z')
  ax.set_title(f'a={a}, b={b}, c={c}')
  plt.show()
```

Задание 5. Нарисовать бифуркационное дерево для одномерного логистического отображения x=rx(1-x); r изменять от 2.4 до 4.0

Бифуркационное дерево - карта, по оси Ox отложен параметр (r), по оси Oy переменная (x); точками отмечается последняя тысяча точек из траектории длинной в 10 тысяч при каждом значении параметра

```python
import matplotlib.pyplot as plt
import numpy as np

def determination_of_next_points(r, x):
    res = r * x * (1 - x)
    return res

iterations = 10000
last_iterations = 1000
x = 0.5
r_values = np.arange(2.4, 4.0, 0.01)
result = []  # для r

for r in r_values:
    x = 0.5  # Сбрасываем x для каждого нового r
    for i in range(iterations):
        x = determination_of_next_points(r, x)

    cur_res = []
    for j in range(last_iterations):
        x = determination_of_next_points(r, x)
        cur_res.append(x)

    result.append(cur_res)
plt.figure(figsize=(12, 8))

for i in range(len(r_values)):
    plt.scatter([r_values[i]] * 1000, result[i],s=2)

plt.title('Бифуркационное дерево ')
plt.xlabel('r')
plt.ylabel('x')
plt.grid()
plt.show()
```
